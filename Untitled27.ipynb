```python
#Install all necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, mean_absolute_error
!pip install pmdarima
from pmdarima import auto_arima
import warnings
warnings.filterwarnings('ignore')
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('stores_sales_forecasting.csv', encoding='latin1') #Upload the dataset
```

```python
data.isna() #Check if there are null values
```

```python
data_1 = data.dropna() #drop null values
```

```python
data_1.shape #To check whether the dimensions of the dataset changed or not
```

```python
data_1
```

```python
df = data_1.copy()
df['Order Date']= pd.to_datetime(df['Order Date']) #Convert the time period into date time order
```

```python
daily_sales = df.groupby('Order Date')['Sales'].sum().reset_index() #Sum sales on daily basis
daily_sales.set_index('Order Date', inplace=True)
```

```python
monthly_sales = daily_sales.resample('M').sum() #sum sales on motnhly basis
```

```python
#Plot the monthly sales across time period
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, label='Monthly Sales', marker='o', linestyle='-')
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
#Check the trend, seasonality and residual distribution of the data
result = seasonal_decompose(monthly_sales, model='additive')
result.plot()
plt.show()
```

It is clearly visible that the sales have an upward trend indicating growth in sales over time. Also, it has a seasonal pattern with periodic peaks and dips. Thus, SARIMA model would be preferntial.


```python
#Check for stationarity
adf_test = adfuller(monthly_sales['Sales'])
print('ADF Statistic:', adf_test[0])
print('p-value:', adf_test[1])
```

Since p value < 0.05, we rejct null hypothesis. Thus, the data is stationary.

```python
#Plot ACF and PACF to find the number of lags for AR(p) and MA(q)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
# Reduced lags to be less than 50% of the sample size
plot_acf(monthly_sales['Sales'].dropna(), lags=20, ax=plt.gca())
plt.title('Autocorrelation Function (ACF)')
plt.xlabel('Lag')
plt.ylabel('Correlation')
plt.show()

plt.subplot(1, 2, 2)
# Reduced lags to be less than 50% of the sample size
plot_pacf(monthly_sales['Sales'].dropna(), lags=20, ax=plt.gca())
plt.title('Partial Autocorrelation Function (PACF)')
plt.xlabel('Lag')
plt.ylabel('Correlation')
plt.show()
```

There are no significant lags except for 0 and 12. Since its monthly data, 12 means 1 years, Thus, we can use (0,0,0) or (1,0,1)

```python
# Find best SARIMA order
auto_model = auto_arima(monthly_sales['Sales'],
                        seasonal=True, m=12,
                        trace=True,
                        stepwise=True,
                        suppress_warnings=True)

# Print best parameters
print(auto_model.summary())
```

Since, ARIMA(0,0,0)(2,1,0)[12] has lowest AIC, it is the best amongst all.

```python
# Fit SARIMA with the best parameters
sarima_model = SARIMAX(monthly_sales['Sales'], order=(0,0,0),seasonal_order=(2,1,0,12))
sarima_fit = sarima_model.fit()
print(sarima_fit.summary())
```

```python
# Forecast next 12 months
forecast_steps = 12
future_dates = pd.date_range(start=monthly_sales.index[-1], periods=forecast_steps+1, freq='M')[1:]
forecast = sarima_fit.get_forecast(steps=forecast_steps)
forecast_ci = forecast.conf_int()

# Plot forecast
plt.figure(figsize=(12,6))
plt.plot(monthly_sales, label='Actual Sales')
plt.plot(future_dates, forecast.predicted_mean, label='Forecast', color='red')
plt.fill_between(future_dates, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='pink', alpha=0.3)
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('SARIMA Forecast for Next 12 Months')
plt.legend()
plt.show()
```

Since, this does not tell much about the accuracy of the model, it would be better to move to train test split method.

```python
df = monthly_sales['Sales']

# Train-test split (80% train, 20% test)
train_size = int(len(df) * 0.8)
train, test = df.iloc[:train_size], df.iloc[train_size:]

# Fit SARIMA model on training data
model = SARIMAX(train, order=(0,0,0), seasonal_order=(1,1,0,12), enforce_stationarity=False, enforce_invertibility=False)
model_fit = model.fit()

# Forecasting on the test set
test_forecast = model_fit.forecast(steps=len(test))

# Future forecast
future_steps = 12
future_forecast = model_fit.forecast(steps=len(test) + future_steps)

# Plot Train, Test, and Forecast
plt.figure(figsize=(12,6))
plt.plot(train.index, train, label="Train Data", color="blue")
plt.plot(test.index, test, label="Test Data (Actual)", color="green")
plt.plot(test.index, test_forecast, label="Test Forecast", color="red", linestyle="dashed")
plt.plot(pd.date_range(test.index[-1], periods=future_steps, freq="MS"),
         future_forecast[len(test):], label="Future Forecast", color="orange", linestyle="dotted")

plt.axvline(test.index[0], color='black', linestyle="dotted", label="Train-Test Split")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("SARIMA Model: Train vs Test vs Forecast")
plt.legend()
plt.show()

# Evaluate model accuracy
mae = mean_absolute_error(test, test_forecast)
rmse = np.sqrt(mean_squared_error(test, test_forecast))

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")
```

The model captured the seasonal component well but still there are some fluctuations uncaptured. Thus, Check the residuals.

```python
import statsmodels.api as sm

# To check the residuals
sarima_fit.plot_diagnostics(figsize=(12, 6))
plt.show()
```

The residuals are random as they show no clear pattern and are stationary also as there are no significant spikes in correlogram,, but a bit skewed which can be the reason of difference between Test Forecast and Actual Test Data.

```python
df = monthly_sales.copy()
df['Sales_log'] = np.log1p(df['Sales'])  # Log transform to stabilize variance

# Train-test split (80% train, 20% test)
train_size = int(len(df) * 0.8)
train, test = df['Sales_log'].iloc[:train_size], df['Sales_log'].iloc[train_size:]

# Fit SARIMAX model on training data
model = SARIMAX(train, order=(1,1,1), seasonal_order=(2,1,1,12), enforce_stationarity=False, enforce_invertibility=False)
model_fit = model.fit()


# Forecasting on the test set
test_forecast = model_fit.forecast(steps=len(test))

# Future forecast
future_steps = 12
future_forecast = model_fit.forecast(steps=len(test) + future_steps)

# Plot Train, Test, and Forecast
plt.figure(figsize=(12,6))
plt.plot(train.index, train, label="Train Data", color="blue")
plt.plot(test.index, test, label="Test Data (Actual)", color="green")
plt.plot(test.index, test_forecast, label="Test Forecast", color="red", linestyle="dashed")
plt.plot(pd.date_range(test.index[-1], periods=future_steps, freq="MS"),
         future_forecast[len(test):], label="Future Forecast", color="orange", linestyle="dotted")

plt.axvline(test.index[0], color='black', linestyle="dotted", label="Train-Test Split")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("SARIMA Model: Train vs Test vs Forecast")
plt.legend()
plt.show()

# Evaluate model accuracy
mae = mean_absolute_error(test, test_forecast)
rmse = np.sqrt(mean_squared_error(test, test_forecast))

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")
```

Logarithmic function being used to counter the skewness, but still there seems need of improvement. Thus, moving towards rolling forecast which is an iterative method would be better.

```python
# Train-Test Split (80-20)
train_size = int(len(df) * 0.8)
train, test = df[:train_size], df[train_size:]

# Find Best SARIMA Order
auto_model = auto_arima(train['Sales'], seasonal=True, m=12,
                        stepwise=True, suppress_warnings=True,
                        error_action="ignore", approximation=False)

best_order = auto_model.order
best_seasonal_order = auto_model.seasonal_order
print(f"Best SARIMA Order: {best_order}, Seasonal: {best_seasonal_order}")

# Rolling Forecast with Optimized SARIMA
predictions = []
history = list(train['Sales'])

for t in range(len(test)):
    # Fit SARIMA with best parameters
    model = SARIMAX(history, order=best_order,
                    seasonal_order=best_seasonal_order,
                    enforce_stationarity=False, enforce_invertibility=False)
    model_fit = model.fit(disp=False)

    # Forecast next step
    yhat = model_fit.forecast()[0]
    predictions.append(yhat)

    # Append actual value to history
    history.append(test['Sales'].iloc[t])

# Future Forecast (Next 12 Steps)
future_steps = 12
future_forecast = model_fit.forecast(steps=future_steps)

# Evaluate Model
mae = mean_absolute_error(test['Sales'], predictions)
rmse = np.sqrt(mean_squared_error(test['Sales'], predictions))
print(f"Mean Absolute Error (MAE): {mae:.3f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.3f}")

# **Plot Results**
plt.figure(figsize=(12, 6))
plt.plot(train.index, train['Sales'], label="Train Data", color='blue')
plt.plot(test.index, test['Sales'], label="Test Data (Actual)", color='green')
plt.plot(test.index, predictions, label="Rolling Forecast", linestyle='dashed', color='red')
plt.plot(pd.date_range(test.index[-1], periods=future_steps, freq="MS"),
         future_forecast, label="Future Forecast", color="orange", linestyle="dotted")
plt.axvline(x=test.index[0], color='k', linestyle='dotted', linewidth=2, label="Train-Test Split")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Optimized SARIMA Rolling Forecast Performance")
plt.legend()
plt.grid(True)
plt.show()
```

The Rolling Forecast method has improved the model accuracy significantly as the measures such as MAE and RMSE have improved. THe model can be further tested by taking logarithmic functions to counter skewness in residuals.
